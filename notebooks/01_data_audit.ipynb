## Data Types and Basic Column Profiling

In this section, we inspect:
- data types
- non-null counts
- null counts
- null percentages
- number of unique values

This gives us a first structured view of the dataset and helps identify columns that may need closer inspection later.

```python
profile_df = pd.DataFrame({
    "column": df.columns,
    "dtype": df.dtypes.values,
    "non_null_count": df.notna().sum().values,
    "null_count": df.isna().sum().values,
    "null_pct": (df.isna().mean().values * 100).round(2),
    "n_unique": df.nunique(dropna=False).values
})

profile_df.sort_values(by="null_count", ascending=False)
```

```python
profile_df[profile_df["null_count"] > 0].sort_values(by="null_count", ascending=False)
```

```python
profile_df.sort_values(by="n_unique", ascending=False).head(15)
```

## Initial Profiling Interpretation

The profiling table reveals several important early patterns in the dataset.

### Observations
- Most columns do not contain missing values.
- Missingness is concentrated in a small number of columns, especially `Churn Reason`, `Churn Category`, `Offer`, and `Internet Type`.
- Some columns have very high cardinality, such as `Customer ID`, `Total Charges`, `Total Revenue`, and location-related fields.
- Several columns have only a small number of unique values, which suggests that the dataset contains many binary or categorical variables.


## Duplicate Check

Before moving further, we check whether the dataset contains fully duplicated rows.

This is important because duplicate records can distort both analysis and model training.

```python
print("Duplicate rows:", df.duplicated().sum())
```

## Duplicate Check Interpretation

The dataset does not appear to contain fully duplicated rows.

This is a positive early signal for data quality, since exact duplicates can distort both descriptive analysis and predictive modeling.

At this stage, no further action is needed regarding row-level duplication.

## Column Type Overview

Here we separate columns into:
- categorical columns
- numeric columns

This is an early structural step that will later help us with EDA, preprocessing, and modeling.

```python
categorical_cols = df.select_dtypes(include=["object"]).columns.tolist()
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()

print("Number of categorical columns:", len(categorical_cols))
print("Number of numeric columns:", len(numeric_cols))

print("\nCategorical columns:")
print(categorical_cols)

print("\nNumeric columns:")
print(numeric_cols)
```

## Column Type Interpretation

The dataset contains a mix of categorical and numeric columns:
- 31 categorical columns
- 19 numeric columns

This suggests that the dataset is structurally suitable for a classification workflow, but it will require separate handling of categorical and numeric features during preprocessing.

At this stage, this separation is only structural and does not yet determine which columns will be used for modeling.

## Initial Inspection of Churn-Related Columns

At this stage, we do not decide which column is the final target.

We only inspect columns whose names suggest a connection to churn, status, or customer identity.
This helps us identify:
- possible target columns
- possible identifier columns
- possible post-outcome or leakage columns

```python
suspect_cols = ["Customer ID", "Customer Status", "Churn Label", "Churn Category", "Churn Reason", "Churn Score"]

for col in suspect_cols:
    print(f"\n--- {col} ---")
    display(df[col].value_counts(dropna=False).head(10))
```

## Churn-Related Columns Interpretation

The dataset contains several columns directly related to churn or customer status, including `Churn Label`, `Customer Status`, `Churn Category`, `Churn Reason`, and `Churn Score`.

### Early observations
- `Churn Label` appears to be a binary field and may be a candidate target variable.
- `Customer Status` appears to contain broader customer outcome categories.
- `Churn Category` and `Churn Reason` contain substantial missingness and seem to be available only for a subset of customers.
- `Churn Score` may represent a derived or model-generated quantity rather than a raw business feature.

### Important note
These columns require deeper inspection before any modeling decision is made, especially to determine whether some of them may introduce leakage or post-outcome information.
